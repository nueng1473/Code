```python
def caculator(num0, num1, op):
    if op == '+': return num0 + num1
    elif op == '-': return num0 - num1
    elif op == '*': return num0 * num1
    else: return num0 / num1
```

```python
caculator(10, 2, '-')
```

```python
caculator(10, 2, '+')
```

```python
caculator(10, 2, '*')
```

```python
caculator(10, 2, '')
```
